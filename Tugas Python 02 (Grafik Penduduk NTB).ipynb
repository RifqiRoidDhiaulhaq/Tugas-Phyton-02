```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
plt.plot(iris['2014'], label='Tahun 2014')
plt.plot(iris['2015'], label='Tahun 2015')
plt.plot(iris['2016'], label='Tahun 2016')
plt.plot(iris['2017'], label='Tahun 2017')
plt.plot(iris['2018'], label='Tahun 2018')
plt.plot(iris['2019'], label='Tahun 2019')
plt.plot(iris['2020'], label='Tahun 2020')

plt.xlabel('Jumlah Data')
plt.ylabel('Jumlah Penduduk')
plt.title('Grafik Jumlah Penduduk Nusa Tenggara Barat Tahun 2014 Sampai Tahun 2020')
plt.legend()
plt.show()
```
