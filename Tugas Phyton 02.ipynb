```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
plt.plot(iris['Lombok Barat'], label='Lombok Barat')
plt.plot(iris['Lombok Tengah'], label='Lombok Tengah')
plt.plot(iris['Lombok Timur'], label='Lombok Timur')
plt.plot(iris['Sumbawa'], label='Sumbawa')
plt.plot(iris['Dompu'], label='Dompu')
plt.plot(iris['Bima'], label='Bima')
plt.plot(iris['Sumbawa Barat'], label='Sumbawa Barat')
plt.plot(iris['Lombok Utara'], label='Lombok Utara')
plt.plot(iris['Kota Mataram'], label='Kota Mataram')

plt.xlabel('Jumlah Data')
plt.ylabel('Jumlah Penduduk')
plt.title('Grafik Jumlah Penduduk Nusa Tenggara Barat Tahun 2014 Sampai Tahun 2020')
plt.legend()
plt.show()
```
